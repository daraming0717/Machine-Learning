## Grouping, Filtering

```python
#library import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# weather 데이터 읽어오기
file_path = 'C:/Users/AtiVer/Documents/ML_gyu/data/weather.csv'
weather = pd.read_csv(file_path, encoding = 'CP949')
weather.head()
```

```python
# 연도, 월, 일 쪼갠 후, '연도', '월', '일' 컬럼 추가
weather['연도'] = pd.DatetimeIndex(weather['일시']).year
weather['월'] = pd.DatetimeIndex(weather['일시']).month
weather['일'] = pd.DatetimeIndex(weather['일시']).day
weather
```

# 그룹핑

```python
# 10년간 연간 최대 평균기온, 최대풍속, 평균풍속을 구하시오.
# max(), proupby()
year_max = weather.groupby('연도').max(numeric_only=True)
year_max
```

```python
#일시, 월, 일 제거
tmp1 = year_max.drop('월',axis = 1)
tmp1
```

```python
tmp2 = tmp1.drop('일',axis = 1)
tmp2
```

```python
year_max = tmp2
year_max
```

```python
#10년간 연간 평균 평균기온, 최대풍속, 평균풍속을 구하시오.
year_mean = weather.groupby('연도').mean(numeric_only = True)
```

```python
tmp1 = year_mean.drop('월',axis = 1)
tmp1
```

```python
tmp2 = tmp1.drop('일',axis = 1)
tmp2
```

```python
year_mean = tmp2
year_mean
```

# pivot()
## 테이블의 구조를 변경 (인덱스, 컬럼을 변경)

```python
#딕셔너리로 데이터프레임 만들기
df = pd.DataFrame({'상품':['시계', '반지', '반지', '목걸이', '팔찌'],
                   '재질':['금', '은', '백금', '금', '은'],
                   '가격':[50000, 20000, 35000, 30000, 60000]})
df
```

```python
#인덱스를 상품으로 변경, 컬럼을 '재질'로 변경
new_df = df.pivot(index = '상품', columns = '재질', values = '가격')
new_df
```

```python
#NaN 값을 0으로 채우기
new_df.fillna(0)
```

# 두 개의 프레임을 연결
## concat() ==> 결합 concatenate
## merge() ==> 고유값을 기준으로 병합 (join)
### merge 방법 : inner, outer, left, right

```python
# concat()
df1 = pd.DataFrame({'A':['a10', 'a11', 'a12'],
                    'B':['b10', 'b11', 'b12'],
                    'C':['c10', 'c11', 'c12']},
                    index = ['가', '나', '다'])

df2 = pd.DataFrame({'B':['b10', 'b11', 'b12'],
                    'C':['c10', 'c11', 'c12'],
                    'D':['d10', 'd11', 'd12']},
                    index = ['다', '라', '마'])
```

```python
df1, df2
```

```python
# concat(df_list, axis = 0, join = 'outer')
# axis:0 => 행방향으로 연결, axis:1 => 열방향으로 연결
# join='outer' => 합집합 생성, join='inner' => 교집합 생성
# 기본값(default) => axis = 0, join = 'outer'
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
# 열의 합집합
df4 = pd.concat([df1, df2], join = 'outer')
df4
```

```python
# 열의 교집합
df5 = pd.concat([df1, df2], join = 'inner')
df5
```

```python
# 행의 합집합, 열은 그대로 결합, axis = 1
df6 = pd.concat([df1, df2], axis = 1, join = 'outer')
df6
```

```python
# 행의 교집합, 열은 그대로 결합, axis = 1
df6 = pd.concat([df1, df2], axis = 1, join = 'inner')
df6
```

```python
# merge
df7 = df1.merge(df2, how = 'outer', on = 'B')
df7
```

```python
df8 = df1.merge(df2, how = 'inner', on = 'B')
df8
```

```python
df8 = df1.merge(df2, how ='outer', left_index = True, right_index = True)
df8
```
