```python
# p128
# x: 0 ~ 50 사이의 값, Y: 50개의 임의의 수 생성
# plot()으로 그래프 생성
n = 50
x = np.arange(50)
y = np.random.random(size=n)

plt.plot(x, y, 'b*:')
```

```python
x = np.linspace(0, np.pi * 2, 100)

# 그래프 사진을 추출하려면 그림 객체를 생성해야함
fig = plt.figure()

plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'b:')
fig.savefig('sin_cos.jpeg')
```

```python
# p130: x = np.linspace(0, np.pi * 2, 100)

# 그래프 사진을 추출하려면 그림 객체를 생성해야함
fig = plt.figure()
plt.title('Sine cosine Curve')

plt.plot(x, np.sin(x), 'r-', label = 'sine curve')
plt.plot(x, np.cos(x), 'b:', label = 'cosine curve')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend(loc = 'upper right')
#fig.savefig('sin_cos.jpeg')
```

```python
# 여러 개의 그래프를 여러 평면으로: subplots()
fig, ax = plt. subplots(2, 2)

X = np.random.randn(100)
Y = np.random.randn(100)
ax[0, 0].scatter(X, Y)

X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
ax[0, 1].bar(X, Y)

X = np.linspace(0, 10, 100)
Y = np.cos(x)
ax[1, 0].plot(X, Y)

Z = np.random.uniform(0, 1, (5, 5))
ax[1, 1].imshow(Z)
```

```python
# subplots()
fig, ax = plt.subplots(2,3)

# wspace: 너비 여백, hspace: 수평 여백
grid = plt.GridSpec(2, 3, wspace = 0.4, hspace = 0.3)
X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0, 0]).scatter(X,Y)

X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0, 1]).bar(X,Y)

X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[1, 1]).plot(X,Y)

X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[1, 2]).scatter(X,Y)
```
