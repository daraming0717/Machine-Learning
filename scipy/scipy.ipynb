```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# p.136
N = 30   # 전체 점의 개수
x = np.random.rand(N)
y = np.random.rand(N)
# print(x,y)

colors = np.random.rand(N)
area = (30 * np.random.rand(N)**2)

plt.scatter(x, y, s = area, c = colors, alpha = 0.5) # 위치, 반지름, 색, 투명도
```

```python
# p.137 : 제주도를 방문한 내국인과 외국인 통계 자료
x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic)
plt.bar(x, foreign)
plt.xticks(x, years)
```

```python
# p.137
x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic, width = 0.25)
plt.bar(x + 0.3, foreign, width = 0.25) # 0.3 만큼 오른쪽으로 이동
plt.xticks(x, years)
```

```python
# p.137
y = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.barh(y, domestic, height = 0.25)
plt.barh(y - 0.3, foreign, height = 0.25) # 0.3 만큼 아래쪽으로 이동
plt.yticks(y, years)
```

```python
# 통계량: 히스토그램, 확률밀도함수, 정규분포, 박스플랏
heights = np.array([175, 165, 164, 164, 171, 165, 160])

plt.hist(heights, bins = 5) # bins = 구간
```

```python
# p.142
# 표준정규분포 : 평균: 0 , 표준편차: 1인 정규분포 N(0,1)
f1 = np.random.randn(100000) # 표준정규분포에서 임의의 난수를 100,000개 샘플링

plt.hist(f1, bins = 200, color = 'red', alpha = 0.7, label ='avg=0, std=1') #f1을 200구간으로 나눠서 빨간색에 투명도 0.7으로 뽑음
plt.axis([-8, 8, -2, 2500]) # x축, y축
plt.legend() # 레이블 출력
```

```python
# p.142
# 표준정규분포 : 평균: 0 , 표준편차: 1인 정규분포 N(0,1)
f1 = np.random.normal(loc = 0,  scale = 1, size = 100000)
f2 = np.random.normal(loc = 3,  scale = 3, size = 100000)

plt.hist(f1, bins = 200, color = 'red', alpha = 0.4, label ='avg=0, std=1')
plt.hist(f2, bins = 200, color = 'green', alpha = 0.4, label ='avg=3, std=3')
plt.axis([-8, 8, -2, 2500])
plt.legend()
```

```python
# SciPy
from scipy.stats import norm
```

```python
# p.144 rvs() : random variable sampling 랜덤 변수 샘플링
# 평균이 0.0이고 분산이 3인 1000개인 데이터를 생성
data = norm.rvs(0.0, 3, size = 1000)

# bins = 5 히스토그램
plt.hist(data, bins = 5, density = True, alpha = 0.4, color = 'r', label = 'bins=5') #density(밀도)
plt.legend()
```

```python
# 1X3 subplots() 이용
fig, ax = plt.subplots(1, 3)
fig.subplots_adjust(wspace = 0.4)
grid = plt.GridSpec(1, 3, wspace = 0.4, hspace = 0.3)

# bins = 5, 평균이 0.0이고 분산이 3인 1000개인 데이터를 생성
data = norm.rvs(0.0, 3, size = 1000)
ax[0].hist(data, bins = 5, density = True, alpha = 0.4, color = 'r', label = 'bins=5')

# bins = 10, 평균이 20.0이고 분산이 3인 1000개인 데이터를 생성
data = norm.rvs(20.0, 3, size = 1000)
ax[1].hist(data, bins = 10, density = True, alpha = 0.4, color = 'g', label = 'bins=10')

# bins = 20, 평균이 40.0이고 분산이 3인 1000개인 데이터를 생성
data = norm.rvs(40.0, 3, size = 1000)
ax[2].hist(data, bins = 20, density = True, alpha = 0.4, color = 'b', label = 'bins=20')
```
